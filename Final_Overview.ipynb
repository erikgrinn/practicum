## Project Goal

The goal of this project is to determine what factors contribute to higher profit for the online store, and which products lead to higher revenue.

## Data Preprocessing and Exploratory Data Analysis

The first stage of any data analysis project involves an overview of the data to ensure that the quality of the information is suitable for analysis. Tasks in this stage include:<br> <br>
-Ensure that the data types are appropriate for each variable <br>
-Rename columns for standardized use <br>
-Check for duplicates <br>
-Check for missing values and decide whether to impute according to distributions <br>

Exploratory Data Analysis involves digging into the data to observe trends. Tasks here include: <br>

-Find outliers, determine whether they are erroneous entries, and decide how they may affect analysis <br>
-Determine which variables may be correlated, possibly using a correlation matrix <br>

## Business Analytics

Now, it is possible to go forward and analyze the data with regards to the project goal. This stage is the opporutnity to determine what factors might result in higher profits. It can be useful to brainstorm questions, such as:  <br> 

1.) Does this online store primarily sell a small amount of high priced items, or a large amount of low priced items? <br>

2.) What are the characteristics (i.e item name) of items that tend to sell more often? <br>

3.) Are there seasonal trends that affect which items are sold? <br>

## Methods

There are several methods we can perform to assess these questions. <br>

1.) It is expected that there would be a higher quantity of lower priced items sold. We can find the range of prices, determine a threshold for lower priced items, and then determine the frequency of items purchased. <br>

2.) It may be that certain products sell more due to keywords. We can lemmatize the product names and see if products with certain words sell more often. <br>

3.) Usually, seasonality affects which items are sold throughout the year. Since we have the date of purchase, we can take that factor into account by analyzing distributions over time. 


Some more techniques that can give rise to insights include: <br>

-Group by quanity of items sold, and find the properties of items that sell in large amounts <br>
-Use K-Means clustering to determine properties of revenue <br>
-Use dashboards (Tableau) to visualize the purchases of products over time <br>
-Group by customer to find out whether customers buy certain products together <br>

## Conclusions

After cleaning the data and adressing these preliminary questions, we can start analyzing the data to determine trends. Through the use of Pandas, plots, and dashboards, we can find answers to these questions as well as answers to new questions formed through the process of analysis.
